```python
# Since There Are so many columns which is not efficient
# We will try to reduce these columns
#insted of taking all categorical values in each  column we will take top 10 most repeated values in that column
```

```python
#Top 10 most repeated values in X1
top10x1 = [i for i in df.X1.value_counts().head(10).sort_values(ascending = False).index]
top10x2 = [i for i in df.X2.value_counts().head(10).sort_values(ascending = False).index]
top10x3 = [i for i in df.X3.value_counts().head(10).sort_values(ascending = False).index]
top10x4 = [i for i in df.X4.value_counts().head(10).sort_values(ascending = False).index]
top10x5 = [i for i in df.X5.value_counts().head(10).sort_values(ascending = False).index]
top10x6 = [i for i in df.X6.value_counts().head(10).sort_values(ascending = False).index]

# we should make binary values for each value
l = [top10x1,top10x2,top10x3,top10x4,top10x5,top10x6]
def encode(df,column,top10):
    for label in top10:
        df[column+'_'+label] = np.where(df[column]==label,1,0)
for i in range(6):
    encode(df,df.columns[i],l[i])
    
```

```python
df.drop(['X1','X2','X3','X4','X5','X6'],axis=1,inplace=True)
```
